```python
# First, installing pygame using pip
!pip install pygame

# Then running your original code
import pygame
import random
import math

pygame.init()
screen = pygame.display.set_mode((800, 600))
clock = pygame.time.Clock()

particles = []

def heart(x, y):
    return (x*x + y*y - 1)**3 - x*x*y*y*y

for _ in range(5000):
    x = random.uniform(-1.5, 1.5)
    y = random.uniform(-1.5, 1.5)
    if heart(x, y) <= 0:
        particles.append([x, y])

t = 0
running = True
while running:
    screen.fill((0, 0, 0))
    t += 0.05
    scale = 180 * (1 + 0.08 * math.sin(t))

    for x, y in particles:
        px = int(400 + x * scale)
        py = int(300 - y * scale)
        pygame.draw.circle(screen, (255, 80, 120), (px, py), 2)

    pygame.display.flip()
    clock.tick(60)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
```
